```python
# configurando chatgpt
import getpass
import os
from google.colab import userdata
from langchain.chat_models import init_chat_model
from langchain_core.vectorstores import InMemoryVectorStore
import openai
from openai import OpenAI
from google.colab import userdata

from pathlib import Path
from IPython.display import Audio, display, clear_output, HTML
import time
import pandas as pd
import numpy as np
import contextlib # Added import
import wave # Added import


client = OpenAI(api_key=userdata.get('OPENAI_API_KEY')) # Moved client initialization here


def transcribe_speech(audio_path):
    import ffmpeg

    # Converte para WAV 16 kHz mono (mais estÃ¡vel para STT)
    wav_path = "input_16k_mono.wav"
    (
        ffmpeg
        .input(audio_path)
        .output(wav_path, ac=1, ar=16000, format='wav', loglevel='error')
        .overwrite_output()
        .run()
    )

    with open(wav_path, "rb") as audio_file:
        transcription = client.audio.transcriptions.create(
            model="gpt-4o-mini-transcribe",
            file=audio_file,
            language="pt",  # ğŸ‘ˆ forÃ§a portuguÃªs
            prompt="audio_file",
            temperature=0
        )
    return transcription.text.strip()



def resposta_ia(messages):
  #completions api
  completion = client.chat.completions.create(
      model="gpt-4o-mini",
      messages= messages
  )
  return completion.choices[0].message.content



def generate_speech(text):
  # ğŸ” ajustado para WAV (facilita calcular duraÃ§Ã£o com wave)
  speech_file_path = "speech.wav"
  with client.audio.speech.with_streaming_response.create(
      model="gpt-4o-mini-tts",
      voice="shimmer",
      input=text,
      instructions="""Responda de forma empolgada, simpÃ¡tica e como se vocÃª fosse um assistente da empresa GoodWe
 uma empresa que auxilia o usuÃ¡rio a ver dados sobre consumo de energia na sua casa (em KWh). """,
      # Removed format="wav"
  ) as response:
    response.stream_to_file(speech_file_path)

# Moved play_and_wait_exact here
def play_and_wait_exact(path, buffer_seconds=1.5, max_buffer=3.0):
  """Reproduz o Ã¡udio e espera exatamente sua duraÃ§Ã£o + buffer (â‰¤ 3s)."""
  display(Audio(path, autoplay=True))
  # calcula a duraÃ§Ã£o real (em WAV Ã© simples com wave)
  try:
    if str(path).lower().endswith(".wav"):
      with contextlib.closing(wave.open(path, 'rb')) as wf:
        frames = wf.getnframes()
        fr = wf.getframerate()
        dur = frames / float(fr)
    else:
      dur = 5.0  # fallback para outros formatos
  except Exception:
    dur = 5.0
  buf = min(buffer_seconds, max_buffer)
  time.sleep(dur + buf)
```

```python
from IPython.display import HTML, display
from google.colab import output
import base64

_RECORD_OR_QUIT_JS = r"""
async function recordOrQuitString() {
  // Cria UI
  const host = document.createElement('div');
  host.style.margin = '12px 0';
  host.innerHTML = `
    <button id="rec_start" style="padding:8px 12px;margin-right:8px;">ğŸ™ï¸ Iniciar gravaÃ§Ã£o</button>
    <button id="rec_stop"  style="padding:8px 12px;margin-right:8px;" disabled>â¹ï¸ Parar</button>
    <button id="rec_quit"  style="padding:8px 12px;">âŒ Encerrar conversa</button>
    <span id="rec_status" style="margin-left:12px;color:#555;">Pronto</span>
  `;
  document.body.appendChild(host);

  const startBtn = document.getElementById('rec_start');
  const stopBtn  = document.getElementById('rec_stop');
  const quitBtn  = document.getElementById('rec_quit');
  const statusEl = document.getElementById('rec_status');

  let stream = null;
  let recorder = null;
  let chunks = [];
  let resolved = false;

  function cleanup() {
    try { if (recorder && recorder.state !== "inactive") recorder.stop(); } catch(e){}
    try { if (stream) stream.getTracks().forEach(t => t.stop()); } catch(e){}
    host.remove();
  }

  startBtn.onclick = async () => {
    chunks = [];
    try {
      stream = await navigator.mediaDevices.getUserMedia({ audio: true });
    } catch (e) {
      cleanup();
      return "error";
    }
    recorder = new MediaRecorder(stream);
    recorder.ondataavailable = e => chunks.push(e.data);
    recorder.start();
    startBtn.disabled = true;
    stopBtn.disabled = false;
    statusEl.textContent = 'Gravando...';
  };

  const result = await new Promise((resolve) => {
    stopBtn.onclick = async () => {
      stopBtn.disabled = true;
      statusEl.textContent = 'Processando Ã¡udio...';
      if (recorder && recorder.state !== "inactive") {
        recorder.stop();
      }
      recorder.onstop = async () => {
        if (resolved) return;
        resolved = true;
        try {
          const blob = new Blob(chunks, { type: 'audio/webm' });
          const arrayBuffer = await blob.arrayBuffer();
          const bytes = new Uint8Array(arrayBuffer);
          const b64 = btoa(String.fromCharCode(...bytes));
          statusEl.textContent = 'Pronto';
          cleanup();
          resolve("audio:" + b64);
        } catch(e) {
          cleanup();
          resolve("error");
        }
      };
    };

    quitBtn.onclick = () => {
      if (resolved) return;
      resolved = true;
      cleanup();
      resolve("quit");
    };
  });

  return result;
}
recordOrQuitString();
"""

def record_from_mic_or_quit(filename="input.webm"):
  """
  Abre a UI e retorna:
    - caminho do arquivo salvo (string) se gravou
    - None se o usuÃ¡rio clicou 'Encerrar conversa'
  LanÃ§a RuntimeError em caso de erro real.
  """
  # (sem prints para nÃ£o duplicar textos)
  result = output.eval_js(_RECORD_OR_QUIT_JS)

  if not isinstance(result, str):
    # Ã s vezes o Colab retorna undefined/None; trate como erro temporÃ¡rio
    raise RuntimeError("Falha ao capturar o Ã¡udio (retorno invÃ¡lido). Tente novamente.")

  if result == "quit":
    return None
  if result == "error":
    raise RuntimeError("Falha ao capturar o Ã¡udio (permita o microfone e tente novamente).")
  if result.startswith("audio:"):
    b64 = result.split("audio:", 1)[1]
    audio_bytes = base64.b64decode(b64)
    with open(filename, "wb") as f:
      f.write(audio_bytes)
    return filename

  # Qualquer outra coisa Ã© inesperada
  raise RuntimeError("Evento inesperado na captura de Ã¡udio (retorno desconhecido).")
```
